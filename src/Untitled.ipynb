```python
import matplotlib.pyplot as plt
from scipy import signal
from scipy.io import wavfile
```

```python
def get_spect(path):
    sample_rate, samples = wavfile.read(path)
    frequencies, times, spectrogram = signal.spectrogram(samples, sample_rate)
    return spectrogram
```

```python
plt.imshow(get_spect('../data/sample1.wav'))
get_spect('../data/sample1.wav').shape
```

```python
import glob

dir_Pn_GVA = glob.glob("../data/Pn_GVA_Ca46/*wav")

for directory in dir_Pn_GVA :
    plt.imshow(get_spect(directory))
    plt.show()
```
